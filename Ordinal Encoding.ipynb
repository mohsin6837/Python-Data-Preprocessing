```python
df = pd.read_csv("ordinal.csv")
```

```python
enc = OrdinalEncoder()
enc.fit_transform(df[["Education"]])
```

```python
df[["Education"]] = enc.fit_transform(df[["Education"]])
```

```python
edu = ['PhD', 'Masters', 'Bachelors']
```

```python
enc = OrdinalEncoder(categories=[edu])
enc.fit_transform(df[["Education"]])
```

```python
df[["Education"]] = enc.fit_transform(df[["Education"]])
```
